```python
# count repositories len(pages_repo_list[0]) * len(pages_repo_list) + len(pages_repo_list[len(pages_repo_list) - 1])
repos_name_list = []
languages_name_list = []
none_list = []
for index, page in enumerate(pages_repo_list):
    for repo in page:
        if repo['name'] is not None and repo['language'] is not None:
            repos_name_list.append(repo['name']) 
        if  repo['language'] is not None:
            languages_name_list.append(repo['language'])

        if repo['language'] is None:
            none_list.append(1)


print(repos_name_list)
print(languages_name_list)
print(len(none_list))
```
